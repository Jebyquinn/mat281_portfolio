1. Lea el "total_profit" de todos los meses, muéstrelo usando un gráfico lineal y un gráfico de dispersión.

```python
x={1: 'Ene',2: 'Feb',3: 'Mar',4: 'Abr',5: 'May', 6:'Jun',7: 'Jul',8: 'Ago',9: 'Sept',10: 'Oct', 11:'Nov',12: 'Dic'}
y=df['total_profit']

plt.subplot(1,2,1) 

plt.plot(x.values(),y,label='linea',color='deeppink')
plt.title("Total profit de todos los meses") 
plt.xlabel('Meses') 
plt.ylabel('total profit')
plt.grid() 

plt.subplot(1,2,2) 
plt.plot(x.values(), y, 'o', color = 'teal')
plt.title("Total profit de todos los meses") 
plt.xlabel('Meses') 
plt.ylabel('total profit') 
plt.grid() 
plt.show() 
```

2. Lea todos los datos de ventas de productos y muéstrelos utilizando un gráfico multilínea.

```python
x={1: 'Ene',2: 'Feb',3: 'Mar',4: 'Abr',5: 'May', 6:'Jun',7: 'Jul',8: 'Ago',9: 'Sept',10: 'Oct', 11:'Nov',12: 'Dic'}
y1=df['facecream']
y2=df['facewash']
y3=df['toothpaste']
y4=df['bathingsoap']
y5=df['shampoo']
y6=df['moisturizer']

plt.figure(figsize=(10, 5))

plt.plot(x.values(),y1,label='linea',color='violet')
plt.plot(x.values(),y2,label='linea',color='black')
plt.plot(x.values(),y3,label='linea',color='blueviolet')
plt.plot(x.values(),y4,label='linea',color='lightskyblue')
plt.plot(x.values(),y5,label='linea',color='teal')
plt.plot(x.values(),y6,label='linea',color='deeppink')

plt.legend(['facecream','facewash','toothpaste','bathingsoap','shampoo','moisturizer'])
plt.title("Ventas de productos a lo largo de los meses") 
plt.xlabel('Meses') 
plt.ylabel('Cantidad vendida de productos')
plt.grid()
plt.show()

#No se ve Face Wash porque tiene los mismos valores que Moisturizer
```

3. Lea los datos de ventas de productos de "facecream" y "facewash" y muéstrelos usando el gráfico de barras.

```python
y1=df['facecream']
y2=df['facewash']

x={1: 'Ene',2: 'Feb',3: 'Mar',4: 'Abr',5: 'May', 6:'Jun',7: 'Jul',8: 'Ago',9: 'Sept',10: 'Oct', 11:'Nov',12: 'Dic'}

plt.figure(figsize=(10, 5))

plt.bar(x.values(),y1,0.35,color='blueviolet')
plt.bar(x.values(),y2,0.35,color='turquoise')

plt.legend(['facecream','facewash','toothpaste','bathingsoap','shampoo','moisturizer'])
plt.title("Ventas de Face cream y face wash a lo largo de los meses") 
plt.xlabel('Meses') 
plt.ylabel('Cantidad vendida de productos')
plt.grid()
plt.show()
```

4. Lea todos los datos de ventas de productos y muéstrelos utilizando un gráfico box-plot.

```python
x1=df['facecream']
x2=df['facewash']
x3=df['toothpaste']
x4=df['bathingsoap']
x5=df['shampoo']
x6=df['moisturizer']

fig = plt.figure(figsize=(10, 5))

plt.boxplot([x1,x2,x3,x4,x5,x6],labels=['Face cream','Face wash','Toothpaste','Bathingsoap','Shampoo','Moisturizer'])
plt.title('Productos en venta')
plt.show()
```

5. Calcule los datos de ventas totales del año pasado para cada producto y muéstrelos usando un gráfico circular

```python
x1=df['facecream'].sum()
x2=df['facewash'].sum()
x3=df['toothpaste'].sum()
x4=df['bathingsoap'].sum()
x5=df['shampoo'].sum()
x6=df['moisturizer'].sum()
labels = 'Face cream', 'Face wash', 'Toothpaste', 'Bathing soap', 'Shampoo', 'Moisturizer'
sizes = [x1, x2, x3, x4,x5,x6]
explode = (0, 0, 0, 0, 0, 0)
colors=['deeppink','teal','blueviolet','turquoise','lightskyblue','violet']

plt.subplots(figsize=(8, 8))
plt.pie(sizes,explode=explode,labels=labels,colors=colors,autopct='%1.1f%%',shadow=True,startangle=90)
plt.axis('equal')
plt.title('Ventas totales de cada producto en porcentaje')

plt.show()
```
